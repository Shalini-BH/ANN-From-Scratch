```python
# Number of rows for the triangle
rows = 5

# Outer loop for each row
for i in range(1, rows + 1):
    # Inner loop 1: Print leading spaces
    for j in range(rows - i):
        print(" ", end="")

    # Inner loop 2: Print the stars
    for k in range(i):
        print("*", end="")

    # Move to the next line after each row
    print()
```
